```python
import numpy as np

import pandas as pd
```

```python
df = pd.read_csv('hyg_v38.csv', usecols=['hr','mag','ra','dec'])
df[['hr','mag','ra','dec']].isna().sum()
```

```python
df = df[(df['hr'].isna() != 1) & (df['mag'] < 4.0)]
df.drop(columns=['hr'],inplace=True)
df['ra'] = np.radians(df['ra']*15)
df['dec'] = np.radians(df['dec'])
df.to_csv(path_or_buf='/catalogue.csv',index=False)
```

```python
# Load your clean radians-scaled CSV text block
df = pd.read_csv('/catalogue.csv')

with open("star_catalog.h", "w") as f:
    f.write("// ========================================================\n")
    f.write("// star_catalog.h\n")
    f.write("// Pre-computed 3D Celestial Reference Unit Vectors\n")
    f.write("// Gate: Magnitude <= 4.0 | Radian Projected | RAM/Cache Packed\n")
    f.write("// ========================================================\n\n")
    f.write("#pragma once\n\n")
    f.write("struct CatalogStar {\n    float x;\n    float y;\n    float z;\n};\n\n")
    f.write(f"const int CATALOG_SIZE = {len(df)};\n")
    f.write("const CatalogStar star_catalog[CATALOG_SIZE] = {\n")
    
    for r_rad, d_rad in zip(df['ra'], df['dec']):
        # Standard spherical-to-Cartesian 3D coordinate mapping equations
        x = np.cos(d_rad) * np.cos(r_rad)
        y = np.cos(d_rad) * np.sin(r_rad)
        z = np.sin(d_rad)
        
        f.write(f"    {{ {x:11.7f}f, {y:11.7f}f, {z:11.7f}f }},\n")
        
    f.write("};\n")

print("star_catalog.h successfully written to disk.")
```
